### Covered API classes
This notebook covers examples for the following API class:
- **SupplyChainIoCFeed** (*TCF-0701*)

### Credentials
Credentials are loaded from a local file instead of being written here in plain text.
To learn how to creat the credentials file, see the **Storing and using the credentials** section in the [README file](./README.md)

1. Importing the required class
First, we will import the required API class from the ticloud module.

```python
from ReversingLabs.SDK.ticloud import SupplyChainIoCFeed
from datetime import datetime, timedelta
```

### 2. Loading the credentials
Next, we will load our Spectra Intelligence credentials from the local `ticloud_credentials.json` file.
**NOTE: Instead of doing this step, you can paste your credentials while creating the Python object in the following step.**

```python
import json


CREDENTIALS = json.load(open("ticloud_credentials.json"))
USERNAME = CREDENTIALS.get("username")
PASSWORD = CREDENTIALS.get("password")
USER_AGENT = json.load(open('../user_agent.json'))["user_agent"]
```

### 3. Creating the SupplyChainIoCFeed object
After we create an object from the SupplyChainIoCFeed class, we will be able to use its methods.

```python
ioc = SupplyChainIoCFeed(
    host="https://data.reversinglabs.com",
    username=USERNAME,
    password=PASSWORD,
    user_agent=USER_AGENT
)
```

### 4. Fetching the IoCs with defined time
We will now fetch IoCs by specifying several parameters:
- time will be defined as a Unix timestamp
- time period will be last 12 hours

```python
twelve_hours = datetime.now() - timedelta(hours=12)
timestamp = str(int(twelve_hours.timestamp()))

response = ioc.pull_with_timestamp(
    time_format="timestamp",
    time_value=timestamp
)

print(response.text)
```

### 5. Fetching IoCs by defining relative time
We will now fetch IoCs from the last 3 days. 

```python
response = ioc.pull_with_relative_time(
    unit="days",
    amount=3
)

print(response)
```

### 6. Setting the starting point
We will set the starting point with the 'start' request.

```python
response = ioc.set_start_time(
    time_format="timestamp",
    time_value=timestamp
)

print(response.text)
```

### 7. Pull IoCs from the starting point
Now that we've set the starting point using the `set_start_time` method, we can start pulling IoCs. 

```python
response = ioc.pull_from_feed()

print(response.text)
```
