# Email file reputation
This notebook demonstrates how to use the `email_file_reputation` method from the RLSDK `AdvancedActions` class.
The `email_file_reputation` method does the following:
- Accepts a hash string of an email file.
- If the email file has graphical attachments, the reputation of those attachments will be taken into account when returning the reputation verdict here.
- If at least one attachment is classified as malicious, the returned verdict will be "MALICIOUS".
- If there are no malicious attachments but there are suspicious ones, the returned verdict will be "SUSPICIOUS".

### Used Spectra Intelligence classes
- **AdvancedActions**
- **FileUpload**

### Credentials
Credentials are loaded from a local file instead of being written here in plain text.
To learn how to creat the credentials file, see the **Storing and using the credentials** section in the [README file](./README.md)

```python
import json
import os
from ReversingLabs.SDK.advanced import AdvancedActions
from ReversingLabs.SDK.ticloud import FileUpload


CREDENTIALS = json.load(open("credentials.json"))
USERNAME = CREDENTIALS.get("ticloud").get("username")
PASSWORD = CREDENTIALS.get("ticloud").get("password")

cwd = os.getcwd()
upper_level = os.path.dirname(cwd)
USER_AGENT = json.load(open(os.path.join(upper_level, "user_agent.json")))["user_agent"]

advanced_actions = AdvancedActions(
    host="https://data.reversinglabs.com",
    username=USERNAME,
    password=PASSWORD,
    user_agent=USER_AGENT
)

file_upload = FileUpload(
    host="https://data.reversinglabs.com",
    username=USERNAME,
    password=PASSWORD,
    user_agent=USER_AGENT
)
```

### 1. Obtain an email file with QR code images
First of all, we need an email file that has attachments such as QR code images resolving into malicious URLs.
Since sending emails with malicious attachments to ourselves and others is not safe nor recommended, we will not provide a tutorial on how to procure the required email file for analysis.

### 2. Upload the email file to Spectra Intelligence
After we obtained the email file, we need to upload it to Spectra Intelligence using the `FileUpload` class from the `ticloud` module in RLSDK. This will send our email file to thorough analysis.

**NOTE:** If you do not or can not find an email file that you could use for this scenario, you can try searching Spectra Analyze with the following search options: `filetype:email tag:email-attachment`

```python
file_upload.upload_sample_from_path("/path/to/my/email_file.eml")
```

### 3. Use the email file reputation method
Now that our email file with malicious QR code attachments is analyzed by Spectra Intelligence, we can use the `email_file_reputation` method on it.

```python
reputation = advanced_actions.email_file_reputation("eml_hash_here")

print(reputation)
```

The URL decoded from the QR code attachment is checked for its reputation. If the URL is found **malicious**, the `email_file_reputation` considers the whole email file malicious and returns a classification string saying "MALICIOUS".
